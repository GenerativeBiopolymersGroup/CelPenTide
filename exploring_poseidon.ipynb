```python
import pandas as pd

with open('POSEIDON.csv', encoding='ISO-8859-1') as file:
    lines = file.readlines()


clean_lines = [line for line in lines if line.count(',') == 11]  # если у нас 12 колонок, то должно быть 11 запятых

from io import StringIO
df = pd.read_csv(StringIO(''.join(clean_lines)))

df.head()
```

```python
import pandas as pd
from io import StringIO

data = []
max_fields = 0

with open('POSEIDON.csv', encoding='ISO-8859-1') as file:
    for line in file:
        try:
            # Пробуем разделить строку на поля
            fields = line.strip().split(',')
            max_fields = max(max_fields, len(fields))  # Определяем максимальное количество полей
            data.append(fields)
        except Exception as e:
            print(f"Error processing line: {line}\nError: {e}")

# Дополняем строки до максимального количества полей
for i in range(len(data)):
    if len(data[i]) < max_fields:
        data[i].extend([None] * (max_fields - len(data[i])))

# Создание DataFrame
columns = [f'Column_{i}' for i in range(max_fields)]
df = pd.DataFrame(data, columns=columns)

# Просмотр первых строк
df.head()
```

```python
# Первые 10 строк
print("Первые 10 строк:")
print(df.head(10))

# 10 строк в середине
middle_index = len(df) // 2
print("\n10 строк в середине:")
print(df.iloc[middle_index-5:middle_index+5])

# Последние 10 строк
print("\nПоследние 10 строк:")
print(df.tail(10))
```

```python
import pandas as pd
from io import StringIO

data = []

with open('POSEIDON.csv', encoding='ISO-8859-1') as file:
    for line in file:
        try:
            # Пробуем разделить строку на поля
            fields = line.strip().split(',')
            if len(fields) > 12:
                fields = fields[:12]  # Сохраняем только первые 12 полей
            elif len(fields) < 12:
                fields.extend([None] * (12 - len(fields)))  # Дополняем недостающие поля
            data.append(fields)
        except Exception as e:
            print(f"Error processing line: {line}\nError: {e}")

# Создание DataFrame
df = pd.DataFrame(data, columns=["Peptide", "Cell line", "Cargo", "PubmedID", "Uptake", "Units", "Conc.", "Time", "Temp.", "Method", "Type", "Sequence"])

# Просмотр первых строк
df.head()
```

```python
import pandas as pd

# Загрузка данных из CSV
df = pd.read_csv('1n.csv', encoding='ISO-8859-1')
# Приведение имен столбцов к соответствующим таблицам
df.rename(columns={
    'Name of CPP': 'trivial_name',
    'Amino acid sequence': 'sequence',
    'CPP type': 'type',
    'Mol. Weight (g/mol)': 'mol_weight'
}, inplace=True)

# Удаление нумерации строк
df = df.drop(columns=['No.'])

# Добавление новой колонки 'name'
df['name'] = "Screening of a Cell-Penetrating Peptide Library in Escherichia coli: Relationship between Cell Penetration Efficiency and Cytotoxicity"
# Добавление новой колонки 'name'
df['doi'] = "10.1021/acsomega.8b02348"
# Удаление последней колонки 'Ref.'
df = df.drop(columns=['Ref.'])
# Сохранение обновленного DataFrame обратно в CSV
df.to_csv('1n.csv', index=False)
# Результат
df
```
